```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
response = requests.get("https://www.sewakost.com/jatim/surabaya/")
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
soup.find('a', class_='link-large').text
```

```python
soup.find('li', class_='fields').text
```

```python
soup.find('li', class_='system').text
```

```python
kos_nama = soup.find_all('a', class_='link-large')
```

```python
kos_alamat = soup.find_all('li', class_='fields')
```

```python
kos_harga = soup.find_all('li', class_='system')
```

```python
for each_kos_nama in kos_nama :
    print (each_kos_nama.text)
```

```python
for each_kos_alamat in kos_alamat :
    print (each_kos_alamat.text)
```

```python
for each_kos_harga in kos_harga :
    print (each_kos_harga.text)
```

```python
data_nama = []
for each_kos_nama in kos_nama:
    data_nama.append(each_kos_nama.text)
```

```python
data_nama
```

```python
data_alamat = []
for each_kos_alamat in kos_alamat:
    data_alamat.append(each_kos_alamat.text)
```

```python
data_alamat
```

```python
data_harga = []
for each_kos_harga in kos_harga:
    data_harga.append(each_kos_harga.text)
```

```python
data_harga
```

```python
data = {
    'NAMA KOS': data_nama,
    'ALAMAT': data_alamat,
    'HARGA': data_harga
}
```

```python
pd.DataFrame(data)
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.to_excel("kos.xlsx", index=False)
df = pd.read_excel('kos.xlsx')
df
```
